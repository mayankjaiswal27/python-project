### Analysis of Generated data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ml_generated_price.csv')
```

```python
df.head()
```

```python
location = input("Enter a location to know the price:")

if location in df['Location'].values:
    last_price = df.loc[df['Location'] == location, 'Predicted Prices'].iloc[-1]
    print(f'The price for {location} is: {last_price}')
else:
    print(f'Location {location} not found in the DataFrame.')
```

### Plotting of the generated data

```python
plt.figure(figsize=(15, 8))
sns.barplot(x='Location', y='Predicted Prices', data=df, errorbar=None)
plt.title('Location v/s The Predicted Price')
plt.xlabel('Location', fontsize=15)
plt.ylabel('Predicted Prices', fontsize=15)
plt.show()
```

### Crime Rate Analysis

```python
crime_rate_labels = df['Location']
crime_rate_values = df['Crime Rate']

plt.figure(figsize=(8, 5))
plt.pie(crime_rate_values, labels=crime_rate_labels, autopct='%1.1f%%', startangle=140)
plt.title('Crime Rate Distribution by Location')
plt.axis('equal')
plt.show()
```
